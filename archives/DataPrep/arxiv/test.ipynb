```python
import arxiv
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm import tqdm

max_workers = 10

def download_paper(paper_id):
    try:
        paper = next(arxiv.Client().results(arxiv.Search(id_list=[str(paper_id)])))
        paper.download_pdf(dirpath="./arxiv_reseach_papers/")
        return f"Downloaded {paper_id}"
    except Exception as e:
        return f"Error downloading {paper_id}: {e}"
    
with ThreadPoolExecutor(max_workers=max_workers) as executor:
        # Schedule the first 10 paper downloads
        futures = [executor.submit(download_paper, arxiv_dataset["train"][i]["id"]) for i in range(len(arxiv_dataset["train"])+1)]

        # Use tqdm to display the progress bar
        for future in tqdm(as_completed(futures), total=len(futures)):
            print(future.result())
```

```python
import os

def rename_files(directory):
    # List all files in the directory
    for filename in os.listdir(directory):
        # Check if the file is a PDF
        if filename.endswith(".pdf"):
            # Get the new filename (first 9 characters + .pdf)
            new_filename = filename[:9] + ".pdf"
            # Construct full file paths
            old_file = os.path.join(directory, filename)
            new_file = os.path.join(directory, new_filename)
            # Rename the file
            os.rename(old_file, new_file)
            print(f"Renamed: {old_file} to {new_file}")

# Define the directory containing the PDF files
directory = "arxiv_reseach_papers"

# Call the rename_files function
rename_files(directory)
```
